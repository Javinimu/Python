```python
df_gb = pd.merge(df_gb, df_sub_gb[['Autor', 'Submission']], on = 'Submission', how = 'left')
df_gb
```

```python
df_gb = df_gb.drop_duplicates()
df_gb.reset_index(drop=True, inplace=True)
df_gb
```

```python
df_gb.rename({'Autor_x': 'Autor_post', 'Autor_y': 'Contestaa'}, axis=1, inplace=True)
df_gb
```

```python
df_gb['Reply']=np.where(df_gb['ParentID']==df_gb['Submission'], df_gb['Contestaa'], df_gb['ParentID'])
df_gb.head()
```

```python
df_com = df_gb
df_com = df_com.drop(['Body', 'Dominant_Topic', 'Perc_Contribution', 'Topic_Keywords', 'Score', 'Creacion', 'Submission', 'Contestaa', 'Reply', 'ParentID'], axis=1)
df_com.rename({'Autor_post': 'Usuario', 'ID': 'Codigo'}, axis=1, inplace=True)
df_com
```

```python
dfResult = df_gb.merge(df_com, left_on='ParentID', right_on='Codigo', how='left')
#df3 = pd.merge(df_gb, df_com[['Autor_post', 'ID']], on = 'ID', how = 'left')
dfResult = dfResult.drop_duplicates()
dfResult.reset_index(drop=True, inplace=True)
dfResult
```

```python
dfResult['Usuario']=np.where(dfResult['Contestaa']==dfResult['Reply'], dfResult['Contestaa'], dfResult['Usuario'])
dfResult
```

```python
dfResult.drop(['Codigo', 'Reply', 'Contestaa', 'Topic_Keywords', 'Perc_Contribution', 'Score', 'Creacion', 'Body', 'ID', 'Submission', 'ParentID'], axis=1, inplace=True)
dfResult.columns = ["Tema", "Source", "Target"]
dfResult.dropna(inplace=True)
dfResult.reset_index(drop=True, inplace=True)
```

```python
dfResult
```

```python
cols = list(dfResult)
cols = [cols[-1]] + cols[:-1]
dfResult = dfResult[cols]
```

```python
dfResult.loc[dfResult['Tema'] == 6, 'Tema'] = 'Valoracion'
dfResult.loc[dfResult['Tema'] == 7, 'Tema'] = 'Prestaciones tecnicas'
dfResult.loc[dfResult['Tema'] == 0, 'Tema'] = 'Hardware'
dfResult.loc[dfResult['Tema'] == 1, 'Tema'] = 'Mantenimiento'
dfResult.loc[dfResult['Tema'] == 5, 'Tema'] = 'Versiones'
dfResult.loc[dfResult['Tema'] == 2, 'Tema'] = 'Coleccionismo'
dfResult.loc[dfResult['Tema'] == 3, 'Tema'] = 'Videojuegos'
dfResult.loc[dfResult['Tema'] == 4, 'Tema'] = 'Software'
dfResult.loc[dfResult['Tema'] == 8, 'Tema'] = 'Miscelaneo'
```

```python
dfResult.drop(dfResult.loc[dfResult['Tema']=='Miscelaneo'].index, inplace=True)
```

```python
dfResult.reset_index(drop=True, inplace=True)
```
